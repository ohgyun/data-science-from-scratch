```python
import pandas as pd
```

```python
# 더미 데이터를 어떻게 만들 수 있을까?
# dummy variables = indicator variables
# 각 컬럼의 값을 어떻게 가공할 수 있을까?
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
# map에 딕셔너리를 정의해 새 값을 할당한다
train['Sex_male'] = train.Sex.map({'female':0, 'male': 1})
```

```python
train.head()
```

```python
pd.get_dummies(train.Sex)

# 시리즈의 유니크한 값 중에, 가능한 값을 한 개의 컬럼으로 할당한다
# 대체로 K개의 유니크한 값이 있다면, K-1개의 더미 값이 있으면 된다
# 남/여 값이 있다면, 한 개의 컬럼이 있으면 된다.
# 기본값을 제외하면 되기 때문이다.
```

```python
pd.get_dummies(train.Sex, prefix='Sex').iloc[:, 1:]

# female을 기본값으로 하고, 1개 컬럼만 남겨두면 된다
# prefix를 붙여두면 구별하기 편하다
```

```python
# 유니크한 값이 여러 개라면?
train.Embarked.value_counts()
```

```python
pd.get_dummies(train.Embarked, prefix='Embarked')
```

```python
# 실제론 3개 중 2개만 있으면 모든 정보를 파악하기 충분하다
embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').iloc[:, 1:]
embarked_dummies.head()
```

```python
# 원본 데이터프레임에 더할 땐
train = pd.concat([train, embarked_dummies], axis=1)
```

```python
train.head()
```

```python
# drop_first를 하면 iloc으로 첫 값을 잘라낸 것과 동일한 결과를 만들 수 있다
pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
```
