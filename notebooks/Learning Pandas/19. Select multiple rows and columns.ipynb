```python
import pandas as pd
```

```python
# loc, iloc, ix에 대한 사용법 및 차이점
```

```python
ufo = pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head(3)
```

```python
# loc = location
# 레이블(index, columns)을 사용해 선택한다
ufo.loc[0, :] # row 0, all columns
```

```python
ufo.loc[[0, 1, 2], :]
```

```python
ufo.loc[0:2, :] # loc는 range의 끝값을 모두 포함한다
```

```python
ufo.loc[:, ['City','State']]
```

```python
# 이름이 있어도 range를 적용할 수 있다
ufo.loc[0:2, 'City':'State']
```

```python
# 아래처럼 해도 같은 결과를 얻을 수 있지만, loc가 훨씬 편리하다
ufo.head(3).drop('Time', axis=1)
```

```python
# 로우를 필터링 할 때 이렇게 했었는데,
ufo[ufo.City == 'Oakland'].State
```

```python
# loc를 사용해도 똑같이 할 수 있다
ufo.loc[ufo.City == 'Oakland', 'State']

# 직접 필터링하는 것보다 loc를 사용하는 것을 더 추천한다
# 동작 방식이 다르고, 이건 나중에 다시 설명한다
```

```python
ufo.iloc[0:2, [0, 3]] # iloc는 오른쪽 끝값을 포함하지 않는다
```

```python
# 두 개의 컬럼을 선택하려고 할 때 이렇게 하면 괄호가 너무 많아 헷갈린다
ufo[['City', 'State']]
```

```python
# 추천
ufo.loc[:, ['City', 'State']]
```

```python
# ix로 조회해보자
drinks = pd.read_csv('http://bit.ly/drinksbycountry', index_col='country')
```

```python
drinks.head()
```

```python
# ix는 레이블과 포지션을 동시에 검새할 때 유용하다
drinks.ix['Albania', 0]
```

```python
drinks.ix[1, 'beer_servings']
```

```python
drinks.ix['Albania', 'beer_servings']
```

```python
drinks.ix['Albania':'Andorra', 0:2] # 위치일 경우 오른쪽 끝값을 포함하지 않는다
```

```python
ufo.ix[0:2, 0:2]

# 인덱스가 인티저일 땐, 0:2가 레이블을 검색하는 것처럼 동작해 양 끝값을 포함한다
# 인덱스가 오브젝트일 땐, 0:2는 위치를 의미해서 오른쪽 끝 값을 포함하지 않는다.
# 내부 로직에 따라 혼란스러울 수 있기 때문에, 특정 값을 조회할 땐 loc를 사용할 것을 추천한다
# ix는 loc,iloc가 추가되기 전에 개발된 메서드이기도 하다.
# 레이블과 위치값으로 동시에 조회할 때만 사용하고,
# 나머지 상황에선 loc를 사용하는 것을 추천한다
```
