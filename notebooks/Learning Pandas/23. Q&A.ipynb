```python
# loc/iloc를 왜 다르게 만든거지?
```

```python
# loc = label location based indexing
ufo.loc[0:4, :]
```

```python
# iloc = integer location based indexing
ufo.iloc[0:4, :]
```

```python
# pandas는 numpy를 사용하는데, iloc은 numpy의 슬라이스를 사용하는 방법과 동일하다
ufo.values[0:4, :]
```

```python
# numpy는 python의 range와 동일한 슬라이스 방법을 사용한다
list(range(0, 4))
```

```python
ufo.head()
```

```python
ufo.loc[:, 'City':'State']

# loc에서 양쪽 끝값을 모두 포함하게 한 이유는, 위치가 아니라 레이블 기반이기 때문이다.
# 'City':'State'와 같이 표현했을 때, 양끝 값을 모두 포함하도록 하는 게 더 직관적이다.
```

```python
# 용량이 엄청 큰 파일을 판다스로 처리할 수 있을까? (3기가 이상)
# 샘플링을 해서 볼 수 있을까?
```

```python
# sample()로 샘플링해볼 수 있다
ufo.sample(n=3)
```

```python
ufo.sample(n=3)
# 결과는 매번 바뀐다
```

```python
# random state(랜덤 시드)를 설정하면 같은 결과를 가져올 수 있다
ufo.sample(n=3, random_state=42)
```

```python
# frac를 정의하면 일부 데이터만 가져올 수 있다
ufo.sample(frac=0.75, random_state=99)
```

```python
# 머신러닝에서 사용하려고, 랜덤으로 학습 데이터와 테스트 데이터를 쪼개고 싶다면?
```

```python
train = ufo.sample(frac=0.75, random_state=99)
```

```python
# 트레인 데이터에 속하지 않은 나머지 데이터를 가져온다
test = ufo.loc[~ufo.index.isin(train.index), :]
```
