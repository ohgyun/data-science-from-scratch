```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]
```

```python
# 여러 조건으로 필터링하기
# 1. pandas가 조건의 순서를 알 수 있게 괄호로 조건을 묶어주고,
# 2. & 또는 | 으로 이어준다
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
# 조건으로 반환되는 건, 역시 데이터프레임과 동일한 크기의 시리즈이다
(movies.duration >= 200) & (movies.genre == 'Drama')
```

```python
# 한 컬럼에 여러 조건을 포함하고 싶다면?
# 이렇게 하면 되지만 좀 장황함
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama')]
```

```python
# 시리즈의 isin()을 사용하면 간편함
movies[movies.genre.isin(['Crime', 'Drama'])]
```
