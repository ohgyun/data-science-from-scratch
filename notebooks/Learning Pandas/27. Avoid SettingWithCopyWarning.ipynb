```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
# 빈 값을 찾아보자
movies.content_rating.isnull().sum()
```

```python
# 빈 로우를 확인해보자
movies[movies.content_rating.isnull()]
```

```python
movies.content_rating.value_counts()
```

```python
# NOT RATED를 빈 값으로 할당해보자
movies[movies.content_rating == 'NOT RATED']
```

```python
import numpy as np
movies[movies.content_rating == 'NOT RATED'].content_rating = np.nan

# SettingWithCopyWarning 이다!
```

```python
movies.content_rating.isnull().sum()

# 위 코드는 동작하지 않았다
```

```python
# loc를 사용해 할당해보자
movies.loc[movies.content_rating == 'NOT RATED', 'content_rating'] = np.nan
```

```python
movies.content_rating.isnull().sum()
```

```python
# movies[movies.content_rating == 'NOT RATED'].content_rating 는 두 개의 액션으로 되어 있다.
# 1) movies[movies.content_rating == 'NOT RATED'] - get item
# 2) .content_rating - set item

# pandas는 get item을 할 때, 실제 데이터의 뷰인지 복사본인지를 확정할 수 없다
# 그래서 위 경고는 복사본에 대해 값을 설정하고 있을 수 있기 때문이다.

# loc[]에 할당하는 것은 바로 set item 을 수행한다
```

```python
# 높은 점수를 받은 영화를 다룬다고 해보자
top_movies = movies.loc[movies.star_rating >= 9, :]
```

```python
top_movies
```

```python
top_movies.loc[0, 'duration'] = 150

# 같은 SettingWithCopyWarning가 발생했다.
# 동일하게 loc[]를 사용했는데 왜일까?
```

```python
top_movies
# 근데 실제로 변경되기도 했다
```

```python
# pandas는 top_movies가 실제 데이터프레임의 뷰인지, 복사본인지를 보장할 수 없기 때문이다
```

```python
# 그래서 명확히 copy()로 할당해주는 게 좋다
top_movies = movies.loc[movies.star_rating >= 9, :].copy()
```

```python
top_movies.loc[0, 'duration'] = 150
```

```python
top_movies
```
